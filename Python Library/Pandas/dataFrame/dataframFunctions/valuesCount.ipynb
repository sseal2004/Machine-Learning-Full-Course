```python
#to find the final or qualifier match
# League matches
league_matches = ipl[ipl['MatchNumber'].str.isdigit()].value_counts()

# Final / Qualifier matches
final_qualifier_matches = ipl[~ipl['MatchNumber'].str.isdigit()].value_counts()
```
