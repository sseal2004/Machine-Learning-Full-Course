```python
import numpy as np
import pandas as pd
```

```python


students = pd.DataFrame({
    'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
    'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
    'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
    'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
    'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]
})

students
```

```python
students['name'].dropna()
```

```python
students
```

```python
students.dropna(how='any') # if any missing values found in a row just remove it
```

```python
students.dropna(how='all') #if a row comntains all missing values then only remove it
```

```python
students.dropna(subset=['name'])
```

```python
#remove those missing values which are there either row and column
students.dropna(subset=['name','college'])
```

```python
students.dropna(inplace=True)#permanently remove all nan value
```
