```python
#find the last match played by virat kohli
ipl['all_players'] = ipl['Team1Players'] + ipl['Team2Players']
ipl.head()
```

```python
def did_kohli_played(players_list):
    return 'V Kohli' in players_list
```

```python
did_kohli_played(['V Kohli','R Sharma','S Dhawan']) #calling that function
```

```python
#among the list of players find whether virat kohli is there or not
ipl['all_players'].apply(did_kohli_played)
```

```python
#create a new column
ipl['didKohliPlayed'] = ipl['all_players'].apply(did_kohli_played)
ipl[(ipl['City'] == 'Delhi') & (ipl['didKohliPlayed'] ==True) ]
```

```python
ipl[(ipl['City'] == 'Delhi') & (ipl['didKohliPlayed'] ==True) ].drop_duplicates(subset=['City','didKohliPlayed'],keep="first")#to keep the recent values
```
