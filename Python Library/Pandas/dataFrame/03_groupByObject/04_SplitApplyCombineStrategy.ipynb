```python
#find the number of movies starting with A for each group
def foo(group):
    print(type(group))
    print(group)
    return group
```

```python
genres.apply(foo)
```

```python
def Newfoo(group):
    return group['Series_Title'].str.startswith('A').sum()

genres.apply(Newfoo)
```

```python
#ranking of each movie oin the basis of imdb score
def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movie)
```

```python
#find normalised IMDB rating group wise
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return  group

genres.apply(normal)
```
