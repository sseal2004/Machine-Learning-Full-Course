```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('../dataSets/imdb-top-1000.csv')
movies
```

```python
movies.head(1)
```

```python
generes = movies.groupby('Genre')
```

```python
#Applying builtin agregate function on groupby objects
generes.sum()
```

```python
#find top 3 generes by total earning
#METHOD 1
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head()
```

```python
#METHOD 2
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head()
```

```python
#find the genre with highest avg IMBD rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head()
```

```python
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
movies.groupby('Genre')['IMDB_Rating'].max()
```

```python
#No. of movies done by each actors
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=True)
```
