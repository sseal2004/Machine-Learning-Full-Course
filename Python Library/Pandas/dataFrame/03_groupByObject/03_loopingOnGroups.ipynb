```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('../dataSets/imdb-top-1000.csv')
movies
```

```python
len(movies.groupby('Genre'))
```

```python
movies['Genre'].nunique()
```

```python
genres = movies.groupby('Genre')
genres.first()
```

```python
genres.last()
```

```python
#looping on groups
for group,data in genres:
    print(type(group),type(data))
```

```python
# To find the highest rated movie in each genre
for group,data in genres:
    print(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
```

```python
#to make it look good make it in dataframe
rows = []

for group, data in genres:
    top_movie = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    rows.append(top_movie)

df = pd.concat(rows, ignore_index=True)
df
```
