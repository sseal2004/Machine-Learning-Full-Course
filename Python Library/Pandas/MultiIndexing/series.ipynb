```python
import numpy as np
import pandas as pd
```

MULTIINDEX SERIES

```python
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('cse',2019),('cse',2020),('cse',2021),('cse',2022),('cse',2023)]
multipleIndex = pd.MultiIndex.from_tuples(index_val)
multipleIndex
```

```python
multipleIndex.levels[1]
```

Hierchieal Indexing

```python
s =pd.Series(
    [1,2,3,4,6,7,8,9],
    index=multipleIndex[:8]
)
s
```

```python
s[('cse',2022)]
```

UNSTACK FUNCTION: TO CONVERT A MULTIINDEX SERIES INTO A DATAFRAME

```python
temp = s.groupby(level=[0,1]).sum().unstack()
temp
```

STACK FUNCTION:TO CONVERT A MULTIINDEX DATAFRAME INTO A SERIES

```python
temp.stack()
```
