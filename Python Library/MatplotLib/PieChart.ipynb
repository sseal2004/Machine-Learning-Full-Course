```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#style
plt.style.use('classic')
```

Univariate/Bivariate Analysis
Categorical vs numerical
Use case-To find the contribution on a standard scale

```python
#simple data
data=[23,43,100,20,40]
subjects=['end','science','math','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
#dataset
df = pd.read_csv('./dataSets/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%1.1f%%')
plt.show()
```

```python
#colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%1.1f%%',colors=['blue','red','yellow','green','orange','purple'])
plt.show()
```

TO FOCUS ANY PARTICULAR CATEGORY - use explode

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0])
```

SHAWDOW

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
```
