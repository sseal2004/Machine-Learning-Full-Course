3D Surface PLot (For Loss Function)

```python
#We want to plot Graph
#f(x,y)=x^2+y^2 
```

```python
# Atfirst generate 100 points for both x and y
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
x
```

```python
xx,yy = np.meshgrid(x,y)
print(yy.shape)
print(xx.shape)
```

Example 1

f(x)= 2x + 2y

```python
z= xx*2 + yy*2
z.shape
```

```python
figure = plt.figure(figsize=(5,5))
ax= plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z)
```

Example 2

f(x)=x^2+y^2

```python
z= xx**2 + yy**2
z.shape
```

```python
figure = plt.figure(figsize=(5,5))
ax= plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z) 
```

```python
figure = plt.figure(figsize=(5,5))
ax= plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
figure.colorbar(p)
```

Example 3

z= sin(x)+cos(y)

```python
z= np.sin(xx)+np.cos(yy)
figure = plt.figure(figsize=(5,5))
ax= plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
figure.colorbar(p)
```

Example 4

z= sin(x)+log(y)

```python
z= np.sin(xx)+np.log(yy)
figure = plt.figure(figsize=(5,5))
ax= plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
figure.colorbar(p)
```

Contour Tour

To Represent a 3D graph into 2D

Example 1

```python
z= xx**2 + yy**2#3d
z.shape
figure = plt.figure(figsize=(5,5))
ax= plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
p
```

```python
z= xx**2 + yy**2#2d
z.shape
figure = plt.figure(figsize=(5,5))
ax= plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis') 
p
fig.colorbar(p)
```

FILL COUNTOUR GRAPH

```python
z= xx**2 + yy**2#2d
z.shape
figure = plt.figure(figsize=(5,5))
ax= plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis') 
p
fig.colorbar(p)
```
