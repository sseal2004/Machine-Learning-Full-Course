```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#style
plt.style.use('classic')
```

```python
batters=pd.read_csv('./dataSet/batter (1).csv')
batters.sample(5)
```

3D SCATTER PLOT

```python
fig=plt.figure()
ax = plt.subplot(projection='3d')
```

Using Scatter 3D function

```python
fig=plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],marker='+')
ax.set_title('IPL batsman Analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
```

3D Line Plot

```python
x=[0,1,5]
y=[0,10,13]
z=[0,13,20]

ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100])
```

```python
x=[0,1,5,10]
y=[0,10,13,0]
z=[0,13,20,40]

ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
```
