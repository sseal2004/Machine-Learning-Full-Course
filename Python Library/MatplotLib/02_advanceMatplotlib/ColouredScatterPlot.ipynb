```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#style
plt.style.use('classic')
```

COLOURED SCATTER PLOT

```python
iris = pd.read_csv('./dataSet/iris (1).csv')
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

Replace - replace

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species']) # 3different types 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

COLOUR BAR - pl.colorbar()

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species']) # 3different types 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

OPACITY - alpha

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=0.6) # 3different types 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

SIZE

```python
plt.Figure(figsize=(25,7))# to enlarge the graph
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species']) # 3different types 
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```
