PIE CHART

With custom dataset

```python
df = pd.DataFrame({
    'batsman':['Dhawan','Rohit','Kohli','Dhoni','Pant','Pandya','SKY'],
    'match1':[120,90,67,78,89,34,20],
    'match2':[0,56,45,90,30,41,4],
    'match3':[90,76,85,40,10,28,40]
})
df.head()
```

```python
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%',explode=[0,0,0,0,0,0.1,0])
```

MULTIPLE PIECHART (SUBPLOTS)

```python
df[['match1','match2','match3']].plot(kind='pie',subplots=True,figsize=(15,8))
```
