```python
batsman = pd.read_csv('../MatplotLib/dataSets/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plot
plt.plot(batsman['index'],batsman['V Kohli'])#blue 
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')#rohit
plt.title('Rohit Sharma Vs Virat Kohil Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
#Line style
#plotting multiple plot
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashed')#blue 
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted')#rohit
plt.title('Rohit Sharma Vs Virat Kohil Career Comparison')
plt.xlabel('Season')
plt.ylabel('Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot')#blue 
```

```python
#Marker
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot',marker='.')#blue 
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',marker='+')#rohit

```

```python
#markerSize
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',marker='+',markersize='4')
```

```python
#label
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot',marker='.',label='Virat')#blue 
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',marker='+',label='Rohit')#rohit
plt.legend()
#location
plt.legend(loc='upper left')
```
