```python
#size
tips = sns.load_dataset('tips')
tips
```

plt.scatter() VS plt.plot()

SLOWER TECHNIQUE (USING plt.scatter() function)

Plotting a SCATER PLOTS on SNS

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

With how many person they have came?

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

FASTER TECHNIQUE(USING plt.plot() function)

2D PLOT

```python
#scatterplot using plt.plot
plt.plot(tips['total_bill'],tips['tip'])
```

USING ADDITIONAL PARAMETER TO MAKE IT CHANGE

```python
plt.plot(tips['total_bill'],tips['tip'],'o')
```
