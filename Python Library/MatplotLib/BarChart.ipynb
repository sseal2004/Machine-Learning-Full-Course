```python
#color and label
df = pd.read_csv('./dataSets/batsman_season_record.csv')
df
```

Bar chart

```python
plt.bar(df['batsman'],df['2015'])
```

Overlaping bar chart

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

Side by side bar plots

```python

x = np.arange(df.shape[0])
width = 0.25

plt.bar(x - width, df['2015'], width=width, label='2015')
plt.bar(x,         df['2016'], width=width, label='2016')
plt.bar(x + width, df['2017'], width=width, label='2017')


plt.xticks(x, df['batsman'])   # if df.index contains labels
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparison of Data from 2015 to 2017")
plt.legend()

plt.show()
```
