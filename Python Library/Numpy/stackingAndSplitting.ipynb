```python
import numpy as np
```

```python
#STACKING
```

```python
# Horizontal stacking
# 2 X 4
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
a5
```

```python
np.hstack((a4,a5))#Horizontal Stacking
```

```python
#Vertical stacking
# 4 X 2
np.vstack((a4,a5))
```

```python
#SPLITING
```

```python
#Horizontal Spliting
a4
```

```python
np.hsplit(a4,2)
```

```python
np.vsplit(a5,3)
```
