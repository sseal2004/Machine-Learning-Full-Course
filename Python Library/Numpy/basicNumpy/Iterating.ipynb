```python
import numpy as np
```

```python
a1 = np.arange(12)#1D
a1
```

```python
a2 = np.arange(12).reshape(4,3)#2D
a2
```

```python
a3 = np.arange(27).reshape(3,3,3)#3D
a3
```

```python
#Loop in 1D array
for i in a1:
    print(i)
```

```python
#Loop in 2D array
for i in a2:
    print(i)
```

```python
#Loop in 3D array
for i in a3:
    print(i)
```

```python
for i in np.nditer(a3):# iterating only the elements
    print(i)
```
