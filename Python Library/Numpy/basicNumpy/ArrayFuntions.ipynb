```python
#DOT PRODUCT OF TWO MATRIX such (a X b) . (b X c)
# important  criteria for dot product among the two matrix col of previous matrix must matches with row of the second matrix

a2 = np.arange(12).reshape(3,4)# Matrix 1
a2
```
