```python
#Reshaping consisted of 3 things:
# 1.Reshape
# 2.Transpose
# 3.Ravel
```

```python
a1 = np.arange(10)#Array 1
a2 = np.arange(12).reshape(3,4)#Array 2
a3 = np.arange(8).reshape(2,2,2)#Array 3
```

```python
#reshape
```

```python
#transpose
np.transpose(a2)
# Easy syntax 
# a2.T(Tranpose)
```

```python
#ravel
# It can convert any dimentional array to 1D
print("Initialy:")
print(a2)
print("After raveling:")
a2.ravel()
```
