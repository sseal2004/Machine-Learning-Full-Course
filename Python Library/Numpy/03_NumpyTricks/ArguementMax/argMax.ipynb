```python
import numpy as np
```

```python
# returns the position or index of the maximum element in an array
```

```python
a= np.random.randint(1,100,15)#1D array
a
```

```python
np.argmax(a)
```

```python
b=np.random.randint(1,100,24).reshape(6,4)#2D array
b
```

```python
np.argmax(b)
```

```python
# maximum element index position in row
np.argmax(b,axis=1)
```

```python
# maximum element index position in col
np.argmax(b,axis=0)
```
