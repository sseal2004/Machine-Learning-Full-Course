```python
# again you can convert 2D array 3d
import numpy as np

a = np.arange(6).reshape(2, 3)   # 2D array
print("Print a:\n")
print(a)
print("shape of a:\n",a.shape)

# Add a new axis (convert to 3D)
b = np.expand_dims(a, axis=2)
print("Print b:\n")
print(b)
print("shape of b:\n",b.shape)
```
