```python
import numpy as np
```

```python
a= np.random.randint(1,100,15)
a
```

```python
#find the index position of the array where it is greater than 50
np.where(a>50)
```

```python
#replace all value >50 with 0
# np.where(condition,true,false)
np.where(a>50,0,a)#replace it with 0
```

```python
# Replace all even numbers with 0
np.where(a%2==0,0,a)
```
