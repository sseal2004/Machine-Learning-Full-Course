```python
import numpy as np
a= np.random.randint(1,100,15)
a
```

```python
b=np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a)
```

```python
np.sort(b,axis=1)#sort row wise
```

```python
np.sort(b,axis=0)#sort column wise
```

```python
np.sort(np.sort(b, axis=1), axis=0)#sort for row and column
```

```python
np.sort(a)[::-1]#sorting in descending order
```
