```python
import numpy as np
```

```python
#Calculte the union among the two set
m = np.array([1,2,3,4,5,6])
n = np.array([3,4,5,6,7])
np.union1d(m,n)
```

```python
np.intersect1d(m,n) #intersection
```

```python
#set diff
np.setdiff1d(m,n)
```

```python
#xor 
np.setxor1d(m,n)
```

```python
np.in1d(m,1) #if condition matches it return true
```
