```python
def sigmoid(array):
    return 1/(1+np.exp(-(array)))

a= np.arange(10)
sigmoid(a)
```

```python
#Binary Cross Entropy (BCE)
#Binary Cross Entropy is a loss function used mainly for binary classification problems (output = 0 or 1).

# BCE=−[ylog(p)+(1−y)log(1−p)]
```

```python
y = np.array([1, 0, 1])
p = np.array([0.9, 0.2, 0.8])

bce = -np.mean(y*np.log(p) + (1-y)*np.log(1-p))
print(bce)#logistic regrestion is bce
```
