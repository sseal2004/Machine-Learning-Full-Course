```python
import numpy as np
```

```python
# Working with missing value -> np.nan
a = np.array([1,2,3,4,np.nan,6])
a
```

```python
# doing boolean indexing to find out the missing value
np.isnan(a)
```

```python
#to get that missing value which is returning true
a[np.isnan(a)]
```

```python
#to get all non missing value just add a "NOT OPERATOR"
a[~np.isnan(a)]
```
