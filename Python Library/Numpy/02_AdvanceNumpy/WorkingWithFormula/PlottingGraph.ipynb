```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PLOTTING GRAPH
```

```python
# Plot a graph x=y
x= np.linspace(10,-10,100) # take 100 element between this point 10 to -10 in equidistance
y=x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) #straight line
```

```python
# y = x^2 (parabola)
x = np.linspace(-10,10,100)
y = x**2

plt.plot(x,y)
```

```python
# y = sin x
x = np.linspace(10,-10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
# y = xlog(x)
x = np.linspace(-10,10,100)
y = x * np.log(x)
plt.plot(x,y)
```

```python
#Plot Sigmoid graph
x=np.linspace(-10,10,100)
y=1/(1+np.exp(-x))
plt.plot(x,y)
```
