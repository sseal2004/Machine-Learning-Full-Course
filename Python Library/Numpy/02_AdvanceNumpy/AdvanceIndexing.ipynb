```python
#Boolean Indexing
a= np.random.randint(1,100,24).reshape(6,4)
```

```python
#find all the number greater than 50
a>50
```

```python
a[a>50] # geetting only those numbers number greater than 50
```

```python
# find out even numbers
a%2 == 0
```

```python
a[a%2==0]
```

```python
#find all the number greater than 50 and are even.
a[(a>50)&(a%2==0)]
```

```python
#find all the number not divisible by 7
a[~(a%7==0)]
```
