FACETGRID

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for dataset
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#figure level --> relplot --> displot -->catplot ---> lmplot
sns.catplot(data=tips,x='sex',y='total_bill')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='violin')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='violin',col='day')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='violin',col='day',row='time')
```

```python
g = sns.FacetGrid(data=tips, col="day", row="time")
g.map(sns.scatterplot, "sex", "total_bill")
g.add_legend()
```

Plotting Pairwise Relationship(PairGrid Vs PairPlot)

    PAIRPLOT

```python
iris = px.data.iris()
```

```python
iris
```

```python
sns.pairplot(iris)#example 1
```

```python
sns.pairplot(iris,hue='species')#example 1
```

PAIRGRID

```python
#pairGrid
g=sns.PairGrid(data=iris)
#g.map
g.map(sns.scatterplot)
```

```python
#hue parameter
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
#map_diag -->map_offering
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)#diagonal --> histogram
g.map_offdiag(sns.scatterplot)#non diagonal ---> scatter plot
```

```python
#map diag ---> map upper ---> map lower 
# Diagonal → Histogram
g.map_diag(sns.histplot)

# Upper triangle → KDE plot (warning disabled)
g.map_upper(sns.kdeplot, warn_singular=False)

# Lower triangle → Scatter plot
g.map_lower(sns.scatterplot)

# Add legend
g.add_legend()

plt.show()
```

JOINTGRID VS JOINTPLOT

```python
sns.jointplot(data=tips,x='sex',y='total_bill')#jointplot
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
#joingrid
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.histplot)
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.violinplot)
```
