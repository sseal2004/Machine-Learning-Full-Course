1.HeatMap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for dataset
```

```python
gap=px.data.gapminder()
gap
```

```python
#Plot rectangular coloured matrix data as colour encoded matrix
```

```python
# covert this gap data into wide format
data = gap.pivot(index='country',columns='year',values='lifeExp')
data
```

```python
# now represent this above grid into a heatmap
```

```python
#axes level function
plt.figure(figsize=(15,10))
sns.heatmap(data)
```

```python
#annot 
#just plot this graph for european conutries
gap[gap['continent'] == 'Europe']
```

Add pivot to expand it 

```python
temp = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
temp
```

```python
sns.heatmap(temp)
```

```python
#To show life expectencies of that country in that time just make annotations = TRUE
sns.heatmap(temp,annot=True)
```

```python
#to add gap between the boxes add linewidth
sns.heatmap(temp,annot=True,linewidths=0.5)
```

```python
#to change the colour 
sns.heatmap(temp,annot=True,linewidths=0.5,cmap='viridis')
```

```python
sns.heatmap(temp,annot=True,linewidths=0.5,cmap='summer')
```
