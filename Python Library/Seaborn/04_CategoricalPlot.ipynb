1. CATEGORICAL SCATTER PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for dataset
```

```python
#import dataset
tips=sns.load_dataset('tips')
iris = px.data.iris()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

1. Strip Plot (One Categorical Column and Another non categorical Column)

```python
tips
```

```python
sns.stripplot(data=tips,x='day',y='total_bill')#axes level Function
```

```python
#But actually it should look like this -->jitter
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

2.CatPlot (Same like above graph)

```python
#using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')#Figure Level Function
```

```python
# #jitter
# (To make the points more scatter)
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.4)
```

```python
#hue parameter
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.4,hue='sex')
```

3. Swarm Plot


```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')#using with hue parameter
```

2. CATEGORICAL DISTRIBUTION PLOT

1.Box Plot

 a standardized graphical tool used in statistics to visualize the distribution of numerical data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

```python
sns.boxplot(data=tips,x='sex',y='total_bill')#example 1
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')# axes level function Method 1
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box')# by using catplot METHOD 2---------->figure level
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')#hue parameter
```

```python
#single boxplot ->numerical col
sns.boxplot(data=tips,y='total_bill')
```

2.Violin Plot(Boxplot + Kde Plot)

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
#hue
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex')
```

```python
#side by side distribution
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

CATEGORICAL ESTIMATE PLOT

1.BAR PLOT

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')#hue parameter
```

```python
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)#find outs the minimum data
```

2.Point Plot

```python
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)#find outs the minimum data
```

```python
#to remove error bar use ci=None
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min,ci=None)
```

```python
#point plot
sns.pointplot(data=tips,x='sex',y='total_bill')
```

3.Count Plot

```python
sns.countplot(data=tips,x='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')# on a particular day male and female distribution
```

```python
#faceting using  catplot
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
```
