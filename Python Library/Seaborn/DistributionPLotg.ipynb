1.HISPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for dataset
```

```python
temp = sns.load_dataset('tips')#predefined dataset in seaborn about a restaurant and customer
temp
```

Funtions are:
1. axis_level - hisplot -->kdeplot -->rugplot
2.figure_level- displot

FOR ONE COULMN STUDY 

```python
#plotting using univariate histogram
sns.histplot(data=temp,x='total_bill')
```

```python
#bins parameter
```

```python
sns.displot(data=temp,x='total_bill',kind='hist',bins=2)
```

```python
#It's also possible to visualise the distribution of a categorical variable using the logic of histogram
#Discrete bins are automatically set for categorical variable
temp
```

```python
sns.displot(data=temp,x='day',kind='hist')
```

```python
#hue parameter
sns.displot(data=temp,x='day',kind='hist',hue='sex')
```

To see this graph more clearly (Only Outlines)

```python
#element ->step
sns.displot(data=temp,x='tip',kind='hist',hue='sex',element='step')
```

Most Famous DataSet in Machine Learning (TITANIC DATASET)

```python
titanic = sns.load_dataset('titanic')
titanic
```

Overall pasenger histogram basis on Age limit

```python
sns.displot(data=titanic,x='age',kind='hist',element='step')
```

```python
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
```

SIde by side representation(FACETING USING COL AND ROW --> NOT WORK ON HISPLOTS FUNCTION)

```python
sns.displot(data=titanic,x='age',kind='hist',col='sex',element='step')
```
