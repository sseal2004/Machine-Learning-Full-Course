```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for dataset
```

```python
temp = sns.load_dataset('tips')#predefined dataset in seaborn about a restaurant and customer
temp
```

```python
sns.scatterplot(data=temp, x='total_bill', y='tip')
plt.show()
```

To classify the gender identically use "hue" parameter

```python
sns.scatterplot(data=temp,x='total_bill',y='tip',hue='sex')
```

Use Style Parameter

```python
sns.scatterplot(data=temp,x='total_bill',y='tip',hue='sex',style='time')# to find how many people come for dinner and lunch
```

```python
# relplot is a figure level function
sns.relplot(data=temp, x='total_bill', y='tip')
```

To classify the gender identically use "hue" parameter

```python
sns.relplot(data=temp, x='total_bill', y='tip',hue='sex')
```

Use Style Parameter

```python
sns.relplot(data=temp, x='total_bill', y='tip',hue='sex',style='time')
```

To increase size

```python
sns.relplot(data=temp, x='total_bill', y='tip',hue='sex',style='time',size='size')
```

Line plot

```python
gap=px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country']=='India']
temp_df
```

```python
#aces level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

Using relplot

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

hue--> style

```python
temp_df=gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

Using relplot

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
temp_df=gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent')
```

SIZE means the thickness of the line 

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

FacetPlot ---> figure level function -->work with relplot
it will work with scatterplot and lineplot function

Example 1

```python
#facetplot
import seaborn as sns
import matplotlib.pyplot as plt

temp = sns.load_dataset('tips')

sns.relplot(data=temp, x='total_bill', y='tip', kind='scatter', hue='sex')

plt.show()
```

Example 2

Making a Subplot side by side into two cols

```python
sns.relplot(data=temp, x='total_bill', y='tip', kind='scatter', col='sex')
```

Example 3

Making a Subplot side by side into two rows

```python
sns.relplot(data=temp, x='total_bill', y='tip', kind='scatter', row='sex')
```

Example 4

```python
sns.relplot(data=temp, x='total_bill', y='tip', kind='scatter',col='smoker' ,row='sex')
```

Example 5

```python
sns.relplot(data=temp, x='total_bill', y='tip', kind='scatter',col='smoker' ,row='day')
```
