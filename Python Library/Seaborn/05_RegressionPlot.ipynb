Regression Plot

1. Regplot 
2. Import

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.


1.RegPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for dataset
```

```python
#import dataset
tips=sns.load_dataset('tips')
iris = px.data.iris()
```

```python
sns.regplot(data=tips,x='total_bill',y='tip')#axes level function
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip')#figure level function
```

```python
#hue parameter
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

2. Residplot(Residual Plot) 

 Concept Linear Regression ----> y distance ----> shows the correctness of the predictions from centre x axis

```python
sns.residplot(data=tips,x='total_bill',y='tip')
```
