1.HISPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #for dataset
```

```python
temp = sns.load_dataset('tips')#predefined dataset in seaborn about a restaurant and customer
temp
```

Funtions are:
1. axis_level - hisplot -->kdeplot -->rugplot
2.figure_level- displot

FOR ONE COULMN STUDY 

```python
#plotting using univariate histogram
sns.histplot(data=temp,x='total_bill')
```

```python
#bins parameter
```

```python
sns.displot(data=temp,x='total_bill',kind='hist',bins=2)
```

```python
#It's also possible to visualise the distribution of a categorical variable using the logic of histogram
#Discrete bins are automatically set for categorical variable
temp
```

```python
sns.displot(data=temp,x='day',kind='hist')
```

```python
#hue parameter
sns.displot(data=temp,x='day',kind='hist',hue='sex')
```

To see this graph more clearly (Only Outlines)

```python
#element ->step
sns.displot(data=temp,x='tip',kind='hist',hue='sex',element='step')
```

Most Famous DataSet in Machine Learning (TITANIC DATASET)

2.KDEPLOT

```python
# Rather than using discrete bins , KDE plot smooths the Observation with Gaussian kernels ,producing a continuous density estimate
#Better than hist as it donot miss any values
```

```python
temp
```

```python
sns.kdeplot(data=temp,x='total_bill')#Method 1
```

```python
sns.displot(data=temp,x='total_bill',kind='kde')#MEthod 2
```

```python
# hue -> fill
sns.displot(data=temp,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=temp,x='total_bill',kind='kde',hue='sex',fill=True) #fill parameter
```

3.RugPlot(Not that much useful)

```python
#Plot marginal distributuion by drawing ticks along x and y areas 
#The function is used for complement other plots by showing the location of individual observation in a unobstrusive way
```

```python
sns.displot(data=temp,x='total_bill',kind='kde')
```

```python
sns.kdeplot(data=temp,x='total_bill')
sns.rugplot(data=temp,x='total_bill')
```

Bivariate histogram
1. A Bivariate histogram bins the data within the rectangle that tile the object 
2. and then shows  the count of observation within each rectangle 

```python
sns.histplot(data=temp,x='total_bill',y='tip')#2d histogram 
#METHOD 1
```

```python
sns.displot(data=temp,x='total_bill',y='tip',kind='hist')#2d histogram 
#METHOD 2
```

Bivariate kdeplot(2d GAUSS FUNCTION)

```python
sns.kdeplot(data=temp,x='total_bill',y='tip')# where ever u see more line that means more customer and less line means less customer 
```
